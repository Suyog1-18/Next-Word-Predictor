```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
toknizer=Tokenizer()
```

```python
toknizer.fit_on_texts([text])
```

```python
len(toknizer.word_index)
```

```python
input_sentance=[]

for sentance in text.split('\n'):

  toknized_sentence=toknizer.texts_to_sequences([sentance])[0]

  for i in range(1,len(toknized_sentence)):
    input_sentance.append(toknized_sentence[:i+1])

print(input_sentance)
```

```python
import numpy as np
```

```python
max_len= max([len(i) for i in input_sentance])
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

pad_input_sentance=pad_sequences(input_sentance,maxlen=max_len,padding='pre')
```

```python
pad_input_sentance
```

```python
import pandas as pd
X=pad_input_sentance[:,:-1]
y=X[:,-1]
```

```python
X.shape
# y.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical

y=to_categorical(y,num_classes=int(len(toknizer.word_index)+1))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,Dense,LSTM
```

```python
X.shape
```

```python
y.shape
```

```python
model=Sequential()
model.add(Embedding(267,80,input_length=8))
model.add(LSTM(160))
model.add(Dense(267,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
print(X.shape)
print(y.shape)
```

```python
model.fit(X,y,epochs=100)
```

```python
text_='the people'
for i in range(1,10):

  #tokanizing
  input_=toknizer.texts_to_sequences([text_])[0]

  #padding
  input__=pad_sequences([input_],maxlen=max_len,padding='pre')

  #predict
  prediction=model.predict(input__)

  pos= np.argmax(prediction)

  for word,index in toknizer.word_index.items():
    if(index==pos):
      print(text_+" "+word)
      text_=text_+" "+word
```
